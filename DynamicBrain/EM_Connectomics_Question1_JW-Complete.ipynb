```python
#get all the clean cells, filter for only those with clean axons



#get all the soma predictions
```

```python
#get all the clean cells, filter for only those with clean axons
clean_cells = client.materialize.query_table('proofreading_status_public_release',
                                       filter_in_dict={'status_axon':['clean','extended']})
```

```python
#get all the soma predictions
soma_preds = client.materialize.query_table('aibs_soma_nuc_metamodel_preds_v117')
```

```python
#merge these dataframes to get the clean basket cells
```

```python
#merge these dataframes to get the clean basket cells
merge_table = pd.merge(clean_cells,soma_preds,how='left',on='pt_root_id')
merge_table.query('cell_type=="BC"')
```
